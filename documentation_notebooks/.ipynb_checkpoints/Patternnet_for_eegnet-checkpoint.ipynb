```python
import sys
sys.path.append("../../pathology_example/braindecode_lazy/")
sys.path.append("../")
from braindecode_lazy.datasets.tuh_lazy import TuhLazy
import layers as patlayers
import torch
import braindecode
from braindecode.models.deep4 import Deep4Net
from torch.nn.functional import relu
import numpy as np
import matplotlib.pyplot as plt

from IPython.core.display import display, HTML

%matplotlib inline
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
train_data = TuhLazy("../../pathology_example//data/eval/", n_recordings=None, target="pathological")
# tuple in train_data with train_data[0][0] the input and train_data[0][1] the label?
```

# Create network

```python
# network params
n_chans = 21
n_classes = 2
n_start_chans = 25
input_time_length = 6000
n_chan_factor = 2
final_conv_length = 1
stride_before_pool = True
nonlin = relu
```

```python
# create network
model = Deep4Net(
    n_chans, n_classes,
    n_filters_time=n_start_chans,
    n_filters_spat=n_start_chans,
    input_time_length=input_time_length,
    n_filters_2=int(n_start_chans * n_chan_factor),
    n_filters_3=int(n_start_chans * (n_chan_factor ** 2.0)),
    n_filters_4=int(n_start_chans * (n_chan_factor ** 3.0)),
    final_conv_length=final_conv_length,
    stride_before_pool=stride_before_pool,
    first_nonlin=nonlin,
    later_nonlin=nonlin).create_network()

braindecode.models.util.to_dense_prediction_model(model)
# load params
model.load_state_dict(torch.load('../../pathology_example/rabea_pathology_deep_with_relu_state_dict_0.pt', 
                                 map_location='cpu'))
model.eval()

# tested if the same result as with torch.load('model.pt') with 
# torch.all(torch.eq(result1, result2)) --> true
```

```python
model_easy = []
for layer in model:
    if layer.__class__.__name__ == 'Dropout':
        continue
    elif layer.__class__.__name__ == 'Expression' and layer.expression_fn.__name__ == 'identity':
        continue
    model_easy.append(layer)
    if not (layer.__class__.__name__ in ['Conv2d', 'BatchNorm2d', 'MaxPool2d'] or (layer.__class__.__name__ == 'Expression' and layer.expression_fn.__name__ == 'relu')):
        print(layer)
```

```python
# only use two samples since memory usage too big otherwise
# normal sample
test_data_0 = torch.FloatTensor(train_data[0][0])
# pathological sample
test_data_1 = torch.FloatTensor(train_data[6][0])

print(test_data_0.shape)
print(test_data_1.shape)
test_data_0 = test_data_0.view(1, test_data_0.shape[0],test_data_0.shape[1], 1)
test_data_1 = test_data_1.view(1, test_data_1.shape[0],test_data_1.shape[1], 1)

test_data = torch.cat((test_data_0, test_data_1), dim=0)
print(test_data.shape)

del test_data_0, test_data_1
```

## New forward model

```python
model_easy
```

```python
pattern_layers = []
for layer in model_easy:
    layerclass = layer.__class__.__name__
    if layerclass == 'Conv2d':
        pattern_layers.append(patlayers.PatternConv2d(layer))
    elif layerclass == 'MaxPool2d':
        pattern_layers.append(patlayers.PatternMaxPool2d(layer))
    elif layerclass == 'Expression' and layer.expression_fn.__name__ == 'relu':
        pattern_layers.append(patlayers.PatternReLU())
    elif layerclass == 'BatchNorm2d':
        pattern_layers.append(patlayers.PatternBatchNorm2d(layer))
    elif layerclass == 'Expression' and layer.expression_fn.__name__ == 'identity':
        pattern_layers.append(layer)
    else:
        print('No PatternLayer for layer:', layer)
```

```python
def forward_pass_for_statistics(inp):
    out = model[0](inp)
#     print(out.shape)
    conv_in_outs = []
    for layer in pattern_layers:
        layerclass = layer.__class__.__name__
        if layerclass == 'PatternConv2d':
            layer_in_out = [out]
            out = layer.forward(out)
            if out.__class__.__name__ == 'tuple':
                layer_in_out.extend(out)
                out = out[0]
            else:
                layer_in_out.append(out)
#             print(*[elem.shape for elem in layer_in_out])
            conv_in_outs.append(layer_in_out)
#             print(out.shape)

        elif layerclass == 'PatternMaxPool2d':
            out, _ = layer.forward(out)
#             print(out.shape)

        elif layerclass == 'PatternReLU':
            out, _ = layer.forward(out)
#             print(out.shape)

        elif layerclass == 'PatternBatchNorm2d':
            out = layer.forward(out)
#             print(out.shape)

        else:
            out = layer(out)
            print('Layer class not known:', layerclass)
        
    out = model[-2:](out)
    
    return out, conv_in_outs

def forward_pass_for_signal(inp):
    out = model[0](inp)
#     print(out)
    pool_sizes = []
    pool_inds = []
    relu_inds = []
    for layer in pattern_layers:
#         print(layer)
        layerclass = layer.__class__.__name__
        if layerclass == 'PatternConv2d':
            print(out.shape)
            out = layer.forward(out)
            if out.__class__.__name__ == 'tuple':
                out = out[0]
        elif layerclass == 'PatternMaxPool2d':
            print('Input to pool:', out.shape)
            pool_sizes.append(out.shape)
            out, inds = layer.forward(out)
            print('Output of pool:', out.shape, inds.shape)
            pool_inds.append(inds)
        elif layerclass == 'PatternReLU':
            out, inds = layer.forward(out)
            relu_inds.append(inds)
        elif layerclass == 'PatternBatchNorm2d':
            out = layer.forward(out)
            
#     print(out.shape)
#     out = model[-2:](out)
    
    return out, pool_inds, relu_inds, pool_sizes

def original_forward(inp):
    out = model[0](inp)
#     print(out.shape)
    for layer in model[1:]:
        out = layer(out)
        if layer.__class__.__name__ == 'Dropout' or (layer.__class__.__name__ == 'Expression' and layer.expression_fn.__name__ == 'identity'):
            continue
#         print(out.shape)
        
    return out


def backward_pass(out, pool_inds, relu_inds, pool_sizes):
    
    pool_cnt = 0
    relu_cnt = 0
    # make a copy of out
    signal = torch.tensor(out)
    # go through the layers in backward order
    for layer in pattern_layers[::-1]:
        print(layer)
        layerclass = layer.__class__.__name__
        if layerclass == 'PatternConv2d':
            print("Shape before conv:", signal.shape)
            signal = layer.backward(signal)
            print('Shape after conv:', signal.shape)
        elif layerclass == 'PatternMaxPool2d':
            print("Shape before maxunpool:", signal.shape)
            signal = layer.backward(signal, pool_inds[pool_cnt], pool_sizes[pool_cnt])
            pool_cnt += 1
            print('Shape after maxunpool:', signal.shape)
        elif layerclass == 'PatternReLU':
            signal = layer.backward(signal, relu_inds[relu_cnt])
            relu_cnt += 1
        elif layerclass == 'PatternBatchNorm2d':
            signal = layer.backward(signal)
            
    return signal
```

```python
out_new, conv_in_outs = forward_pass_for_statistics(test_data[:,:,:1000])
print()
out_orig = original_forward(test_data[:,:,:1000])
```

```python
# original forward pass and new defined forward pass the same?
print(torch.equal(out_new, out_orig))
```

```python
# define how long the samples should be; again, whole length does not
# work because memory consumption is too big then
sample_len = 5000
test_inp = test_data[:,:,:sample_len,:]
print(test_data.shape, test_inp.shape)
out_new, conv_in_outs = forward_pass_for_statistics(test_inp)
```

## Compute statistics

```python
conv_cnt = 0 
for layer in pattern_layers:
    layerclass = layer.__class__.__name__
    if layerclass == 'PatternConv2d':
        if layer.forward_layer.dilation != (1,1):
            print(layer.forward_layer)
            print(conv_in_outs[conv_cnt][0].shape, conv_in_outs[conv_cnt][1].shape)
            print(layer.backward_layer)
            print()
        layer.compute_statistics(*conv_in_outs[conv_cnt])
        conv_cnt += 1
```

## Compute patterns

```python
for layer in pattern_layers:
    layerclass = layer.__class__.__name__
    if layerclass == 'PatternConv2d':
        layer.compute_patterns()
        layer.set_patterns()
```

## Compute signal

```python
# compute forward pass
out, pools, relus, pool_sizes = forward_pass_for_signal(test_inp) 
# compute output with only maximum values
out_tmp = torch.zeros(out.size())
max_v, max_i = torch.max(out.data, dim=1)
# did not find out how to use max_i for indexing, but in this example 
# the first sample always has its maximum at the first value,
# the second sample always has its maximum at the second value
out_tmp[0,0,:] = max_v[0]
out_tmp[1,1,:] = max_v[1]
```

```python
# now we have the output prepared, so let's move 
# on to the backward pass
signal = backward_pass(out_tmp, pools[::-1], relus[::-1], pool_sizes[::-1])
```

```python
print(signal.shape)
print(test_inp.shape)
```

```python
sig_min = torch.min(signal).detach()
sig_max = torch.max(signal).detach()
print(sig_min, sig_max)

test_data_norm = test_inp / torch.max(torch.abs(test_inp))
print(torch.max(test_data_norm), torch.min(test_data_norm))

# eeg for normal sample
plt.figure(figsize=(20,15))

for i in range(signal.shape[3]):
    plt.subplot(signal.shape[3],1,i+1)
    
    plt.plot(np.array(test_inp[0,i,:,0]))
    
plt.ylim(sig_min, sig_max)
plt.show()
```
